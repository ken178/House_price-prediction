```python
import pandas as pd
df=pd.read_csv('bengaluru_house_prices.csv')
```

```python
df.shape
```

```python
df = df.drop(['area_type','society','balcony','availability'],axis='columns')
df.shape
```

```python
df.head(5)
```

<h1 style='color:green' align='center'>data cleaning</h2>

```python
df.isnull().sum()
```

```python
df=df.dropna()
df.shape
```

```python
df.isnull().sum()
```

```python
df['size'].value_counts()
```

```python
df['size']=df['size'].str.extract(r'(\d+)')
```

```python
df['size'].dtype
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='area_type', y='price', data=df, estimator=np.mean, ci=None)
plt.xticks(rotation=45)
plt.title('Average Price by Area Type')
plt.show()
```
